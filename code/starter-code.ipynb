##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

```python
import scipy.stats as stats
```

```python
# H0 - The difference in participation rates between the ACT test and SAT test is 0
# H1 - The difference in participation rates between the ACT test and SAT test is not 0
```

```python
print(np.mean(sat_act_merge['SAT Participation %']))
print(np.mean(sat_act_merge['ACT Participation %']))
print(np.mean(sat_act_merge['ACT Participation %']) - np.mean(sat_act_merge['SAT Participation %']))
```

```python
# result = stats.ttest_ind(experimental, control)
# result.statistic, result.pvalue
```

```python
stats.ttest_ind(sat_act_merge['ACT Participation %'], sat_act_merge['SAT Participation %'])
```

```python
t_stat, p_value = stats.ttest_ind(sat_act_merge['ACT Participation %'], sat_act_merge['SAT Participation %'])
t_stat, p_value
```

```python
# alpha = 0.5
# P_value = 0.0002413
# p_value < alpha, therefore there is evidence to reject the null hypothesis
```

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
n = sat_act_merge.shape[0]
act_participation_mean = np.mean(sat_act_merge['ACT Participation %'])
act_participation_std_error = np.std(sat_act_merge['ACT Participation %'],
                                    ddof=1)/np.sqrt(n)
```

```python
stats.norm.interval(0.95, act_participation_mean, act_participation_std_error)
```

```python
n = sat_act_merge.shape[0]
sat_participation_mean = sat_act_merge['SAT Participation %'].mean()
sat_participation_std_error = np.std(sat_act_merge['SAT Participation %'],
                                    ddof=1)/np.sqrt(n)
```

```python
stats.norm.interval(0.95, sat_participation_mean, sat_participation_std_error)
```

# There is a 95% probability that the participation rate of students taking the ACT is between 56.43% and 74.08%
# There is a 95% probability that the participation rate of students taking the SAT is between 30.12% and 49.49%
